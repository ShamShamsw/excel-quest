## Step 1 — Create the Sample Data

We have two tables:
- **Treasure Legend**: maps Chest IDs to their contents and gold value
- **Treasure Map**: a list of chest IDs found on the island (some may be invalid)

````python
# Treasure Legend — the lookup table (equivalent to the second sheet in challenge.xlsx)
legend = pd.DataFrame({
    'ChestID': ['C001', 'C002', 'C003', 'C004', 'C005', 'C006', 'C007', 'C008'],
    'Contents': ['Gold Coins', 'Silver Bars', 'Rubies', 'Emeralds', 'Sapphires',
                 'Ancient Map', 'Diamond Ring', 'Fool\'s Gold'],
    'GoldValue': [500, 300, 750, 600, 800, 1200, 2000, 50]
})

# Treasure Map — the main table with IDs to look up (some are invalid)
treasure_map = pd.DataFrame({
    'Location': ['Skull Rock', 'Dead Man\'s Cove', 'Siren Shore', 'Mermaid Bay',
                 'Cannon Point', 'Shark Reef', 'Ghost Island', 'Parrot Perch'],
    'ChestID': ['C003', 'C001', 'C999', 'C007', 'C005', 'C002', 'C888', 'C008']
})

print('Treasure Legend:')
print(legend.to_string(index=False))
print()
print('Treasure Map:')
print(treasure_map.to_string(index=False))
````

## Step 2 — VLOOKUP Equivalent: Match Chest IDs to Contents

In Excel, the formula would be:
```
=IFERROR(VLOOKUP(A2, Legend!$A$2:$C$9, 2, FALSE), "Not Found")
```

In Python, we use `pd.merge()` with `how='left'` — this is equivalent to a VLOOKUP with IFERROR:
- **Left join** keeps all rows from `treasure_map`
- Unmatched IDs get `NaN`, which we fill with `"Not Found"`

````python
# Perform the VLOOKUP equivalent
result = treasure_map.merge(legend, on='ChestID', how='left')

# Handle missing values — equivalent to IFERROR(..., "Not Found")
result['Contents'] = result['Contents'].fillna('Not Found')
result['GoldValue'] = result['GoldValue'].fillna(0)

print('Lookup Result:')
print(result.to_string(index=False))
````

## Step 3 — Calculate Total Value of Specified Chests

Your mission requires calculating the total gold value for a specific list of chest IDs.

In Excel:
```
=SUMPRODUCT((A2:A9=target_id)*C2:C9)
```

````python
# Specify the chests we want to total
target_chests = ['C003', 'C007', 'C005']

# Filter only valid (found) chests in the target list
found = result[result['ChestID'].isin(target_chests)]

total_value = found['GoldValue'].sum()

print(f'Target chests: {target_chests}')
print()
print(found[['Location', 'ChestID', 'Contents', 'GoldValue']].to_string(index=False))
print()
print(f'Total Gold Value: {total_value} gold coins')
````

## Step 4 — Find the Most Valuable Chest

In Excel:
```
=INDEX(A2:A9, MATCH(MAX(C2:C9), C2:C9, 0))
```

````python
# Find the most valuable chest on the island
max_row = legend.loc[legend['GoldValue'].idxmax()]

print(f'Most Valuable Chest:')
print(f"  Chest ID : {max_row['ChestID']}")
print(f"  Contents : {max_row['Contents']}")
print(f"  Value    : {max_row['GoldValue']} gold coins")
````
