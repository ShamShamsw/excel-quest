We replicate the grouping, summarizing, and slicing that you would do in Excel Pivot Tables.

````python
import pandas as pd
import numpy as np

print('Libraries loaded!')
````

## Step 1 — Generate Sample Sales Data

We create a realistic sales dataset with Region, Product, Date, Sales, and Cost columns.

````python
np.random.seed(42)

regions = ['North', 'South', 'East', 'West']
products = ['Widget A', 'Widget B', 'Gadget X', 'Gadget Y', 'Thingamajig']
dates = pd.date_range('2026-01-01', '2026-06-30', freq='W')  # weekly dates

rows = []
for date in dates:
    for region in regions:
        product = np.random.choice(products)
        sales = np.random.randint(5000, 50000)
        cost = int(sales * np.random.uniform(0.5, 0.8))
        rows.append({'Date': date, 'Region': region, 'Product': product,
                     'Sales': sales, 'Cost': cost})

df = pd.DataFrame(rows)
df['Month'] = df['Date'].dt.to_period('M')
df['Profit'] = df['Sales'] - df['Cost']

print(f'Dataset shape: {df.shape}')
print(df.head(8).to_string(index=False))
````

## Step 2 — Pivot Table: Sales by Region

**Excel equivalent:** Drag `Region` to Rows, `Sales` to Values (Sum)

````python
# Pivot Table: Total Sales by Region
pivot_region = df.pivot_table(
    values='Sales',
    index='Region',
    aggfunc='sum'
).rename(columns={'Sales': 'Total Sales'})

pivot_region['Total Sales'] = pivot_region['Total Sales'].map('${:,.0f}'.format)

print('📊 Pivot Table — Total Sales by Region:')
print(pivot_region.to_string())
````

## Step 3 — Pivot Table: Sales by Product

**Excel equivalent:** Drag `Product` to Rows, `Sales` to Values (Sum)

````python
# Pivot Table: Total Sales by Product
pivot_product = df.pivot_table(
    values='Sales',
    index='Product',
    aggfunc='sum'
).rename(columns={'Sales': 'Total Sales'}).sort_values('Total Sales', ascending=False)

pivot_product['Total Sales'] = pivot_product['Total Sales'].map('${:,.0f}'.format)

print('📊 Pivot Table — Total Sales by Product:')
print(pivot_product.to_string())
````

## Step 4 — Group by Month (Date Grouping)

**Excel equivalent:** Right-click date in PivotTable → Group → Months

````python
# Pivot Table: Sales by Month
pivot_month = df.pivot_table(
    values='Sales',
    index='Month',
    aggfunc='sum'
).rename(columns={'Sales': 'Total Sales'})

pivot_month['Total Sales'] = pivot_month['Total Sales'].map('${:,.0f}'.format)

print('📊 Pivot Table — Total Sales by Month:')
print(pivot_month.to_string())
````

## Step 5 — Slicer Equivalent: Filter by Region

**Excel equivalent:** Insert → Slicer → select Region field

Here we demonstrate filtering the pivot table for a specific region.

````python
# Slicer equivalent: show monthly sales for 'North' region only
selected_region = 'North'

pivot_sliced = df[df['Region'] == selected_region].pivot_table(
    values='Sales',
    index='Month',
    aggfunc='sum'
).rename(columns={'Sales': 'Total Sales'})

pivot_sliced['Total Sales'] = pivot_sliced['Total Sales'].map('${:,.0f}'.format)

print(f'📊 Slicer Active: Region = {selected_region}')
print(pivot_sliced.to_string())
````

## Step 6 — Calculated Field: Profit Margin

**Excel equivalent:** PivotTable → Fields, Items & Sets → Calculated Field
```
Profit Margin = (Sales - Cost) / Sales
```

````python
# Calculated field: Profit Margin by Region
pivot_margin = df.pivot_table(
    values=['Sales', 'Cost', 'Profit'],
    index='Region',
    aggfunc='sum'
)
pivot_margin['Profit Margin %'] = (pivot_margin['Profit'] / pivot_margin['Sales'] * 100).round(1)

print('📊 Calculated Field — Profit Margin by Region:')
print(pivot_margin[['Sales', 'Cost', 'Profit', 'Profit Margin %']].to_string())
````
