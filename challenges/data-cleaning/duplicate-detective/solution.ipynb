````python
import pandas as pd

print('Libraries loaded!')
````

## Step 1 — Create the Messy Crime Database

We simulate the city's crime database with intentional duplicate entries.

````python
# Crime database with duplicates (CaseID C004, C007, C009 appear twice)
data = {
    'CaseID':    ['C001', 'C002', 'C003', 'C004', 'C005',
                  'C006', 'C007', 'C004', 'C008', 'C009',
                  'C007', 'C009', 'C010'],
    'Suspect':   ['Alice', 'Bob', 'Carol', 'Dave', 'Eve',
                  'Frank', 'Grace', 'Dave', 'Hank', 'Ivy',
                  'Grace', 'Ivy', 'Jack'],
    'Crime':     ['Theft', 'Fraud', 'Vandalism', 'Theft', 'Robbery',
                  'Fraud', 'Theft', 'Theft', 'Vandalism', 'Fraud',
                  'Theft', 'Fraud', 'Robbery'],
    'Date':      ['2026-01-05', '2026-01-07', '2026-01-10', '2026-01-12', '2026-01-15',
                  '2026-01-18', '2026-01-20', '2026-01-12', '2026-01-22', '2026-01-25',
                  '2026-01-20', '2026-01-25', '2026-01-28']
}

df = pd.DataFrame(data)

print(f'Original dataset: {len(df)} rows')
print(df.to_string(index=True))
````

## Step 2 — Flag Duplicates with COUNTIF

**Excel formula:**
```
=IF(COUNTIF($A$2:$A$14, A2)>1, "DUPLICATE", "OK")
```

In pandas, `duplicated()` returns `True` for duplicate rows (keeping the first occurrence).
We use `keep=False` to flag ALL occurrences (including the first), matching Excel's COUNTIF > 1 behavior.

````python
# COUNTIF equivalent: flag rows where CaseID appears more than once
df['DuplicateFlag'] = df.duplicated(subset=['CaseID'], keep=False).map({True: 'DUPLICATE', False: 'OK'})

print('Dataset with duplicate flags:')
print(df.to_string(index=True))
print()
print(f"Duplicates found: {(df['DuplicateFlag'] == 'DUPLICATE').sum()} rows")
````

## Step 3 — Highlight Duplicates (Conditional Formatting Equivalent)

In Excel you would use **Home → Conditional Formatting → Highlight Cell Rules → Duplicate Values**.

Here we display duplicates separately to simulate the visual highlight.

````python
# Show only the duplicate rows (equivalent to conditional formatting highlighting them)
duplicates = df[df['DuplicateFlag'] == 'DUPLICATE']

print('⚠️  Highlighted duplicates:')
print(duplicates[['CaseID', 'Suspect', 'Crime', 'Date']].to_string(index=True))
````

## Step 4 — Remove Duplicates

**Excel:** Data tab → Remove Duplicates → select CaseID column → OK

In pandas: `drop_duplicates(subset=['CaseID'], keep='first')` keeps the first occurrence and removes the rest.

````python
# Remove duplicates — keep the first occurrence of each CaseID
df_clean = df.drop_duplicates(subset=['CaseID'], keep='first').drop(columns=['DuplicateFlag'])
df_clean = df_clean.reset_index(drop=True)

print(f'Cleaned dataset: {len(df_clean)} rows (removed {len(df) - len(df_clean)} duplicates)')
print(df_clean.to_string(index=True))
````

## Step 5 — Verify the Clean Dataset

After removing duplicates, run COUNTIF again to confirm no duplicates remain.

````python
# Verify: no CaseID should appear more than once
remaining_dups = df_clean.duplicated(subset=['CaseID'], keep=False).sum()

if remaining_dups == 0:
    print('✅ Verification passed: No duplicate CaseIDs found!')
else:
    print(f'❌ Verification failed: {remaining_dups} duplicate rows still remain.')
````
