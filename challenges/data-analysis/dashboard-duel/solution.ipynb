This notebook demonstrates the dashboard analysis techniques using Python and `matplotlib`.
We replicate the charts, heatmaps, and sparklines you would build in an Excel dashboard.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

print('Libraries loaded!')
````

## Step 1 — Generate the Dashboard Dataset

````python
np.random.seed(21)

products = ['Widget A', 'Widget B', 'Gadget X', 'Gadget Y', 'Thingamajig']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']

# Monthly sales by product
sales_data = {p: np.random.randint(15000, 80000, len(months)) for p in products}
monthly_df = pd.DataFrame(sales_data, index=months)

# Summary: total sales by product
product_totals = monthly_df.sum().sort_values(ascending=False)

print('Monthly Sales by Product:')
print(monthly_df.to_string())
print()
print('Total Sales by Product:')
print(product_totals.map('${:,.0f}'.format).to_string())
````

## Step 2 — Bar Chart: Total Sales by Product

**Excel equivalent:** Select product totals → Insert → Clustered Bar Chart

````python
fig, ax = plt.subplots(figsize=(8, 4))

bars = ax.barh(product_totals.index, product_totals.values, color='steelblue', edgecolor='white')
ax.set_xlabel('Total Sales ($)')
ax.set_title('Total Sales by Product', fontweight='bold')
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x/1000:.0f}K'))

# Add value labels
for bar, val in zip(bars, product_totals.values):
    ax.text(val + 500, bar.get_y() + bar.get_height()/2,
            f'${val/1000:.0f}K', va='center', fontsize=9)

plt.tight_layout()
plt.savefig('/tmp/bar_chart.png', dpi=100, bbox_inches='tight')
plt.show()
print('Bar chart saved to /tmp/bar_chart.png')
````

## Step 3 — Line Chart: Monthly Sales Trends

**Excel equivalent:** Select monthly data → Insert → Line Chart

````python
fig, ax = plt.subplots(figsize=(9, 4))

for product in products:
    ax.plot(months, monthly_df[product], marker='o', label=product, linewidth=2)

ax.set_ylabel('Sales ($)')
ax.set_title('Monthly Sales Trends by Product', fontweight='bold')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x/1000:.0f}K'))
ax.legend(loc='upper left', fontsize=8)
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('/tmp/line_chart.png', dpi=100, bbox_inches='tight')
plt.show()
print('Line chart saved to /tmp/line_chart.png')
````

## Step 4 — Heatmap (Conditional Formatting Color Scale)

**Excel equivalent:** Select data range → Conditional Formatting → Color Scales

````python
fig, ax = plt.subplots(figsize=(8, 4))

heatmap = ax.imshow(monthly_df.values.T, cmap='RdYlGn', aspect='auto')

ax.set_xticks(range(len(months)))
ax.set_xticklabels(months)
ax.set_yticks(range(len(products)))
ax.set_yticklabels(products)
ax.set_title('Sales Heatmap — Conditional Formatting Color Scale', fontweight='bold')

# Add value annotations
for i, product in enumerate(products):
    for j, month in enumerate(months):
        val = monthly_df.loc[month, product]
        ax.text(j, i, f'${val/1000:.0f}K', ha='center', va='center', fontsize=8)

plt.colorbar(heatmap, ax=ax, label='Sales')
plt.tight_layout()
plt.savefig('/tmp/heatmap.png', dpi=100, bbox_inches='tight')
plt.show()
print('Heatmap saved to /tmp/heatmap.png')
````

## Step 5 — Sparklines (In-line Trend Visualization)

**Excel equivalent:** Select a cell → Insert → Sparklines → Line

Here we print ASCII sparklines to show the trend direction for each product.

````python
def ascii_sparkline(values):
    """Create a simple ASCII sparkline from a series of values."""
    bars = '▁▂▃▄▅▆▇█'
    min_v, max_v = min(values), max(values)
    if max_v == min_v:
        return '─' * len(values)
    normalized = [(v - min_v) / (max_v - min_v) for v in values]
    return ''.join(bars[int(n * (len(bars) - 1))] for n in normalized)

print('Sparkline Summary Table:')
print(f'{"Product":<16} {"Sparkline":<12} {"Min":>8} {"Max":>8} {"Total":>10}')
print('-' * 60)
for product in products:
    values = monthly_df[product].tolist()
    spark = ascii_sparkline(values)
    print(f'{product:<16} {spark:<12} ${min(values):>7,.0f} ${max(values):>7,.0f} ${sum(values):>9,.0f}')
````
